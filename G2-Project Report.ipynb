## IV.  Model Implementation

**1. Helper Functions**

```python
# Helper function

def rmse_cv(model):
    rmse = np.sqrt(-cross_val_score(model, X, Y, scoring="neg_mean_squared_error", cv=5)).mean()
    return rmse
    

def evaluate(Y, predictions):
    mae = mean_absolute_error(Y, predictions)
    mse = mean_squared_error(Y, predictions)
    rmse = np.sqrt(mean_squared_error(Y, predictions))
    r_squared = r2_score(Y, predictions)
    return mae, mse, rmse, r_squared

model_evaluation = pd.DataFrame(columns=["Model","MAE","MSE","RMSE","R2 Score","RMSE (Cross-Validation)"])
```

**2. Feature Selection**

- In addition to selecting numerical features with correlation rates above 0.50 or below -0.50 based, we also considered categorical features that we believe have a significant effect on the target variable

```python
num_features = list(df2.corr()["SalePrice"][(df2.corr()["SalePrice"]>0.50) | (df2.corr()["SalePrice"]<-0.50)].index)
cat_features = ["MSZoning", "Utilities","BldgType","Heating","KitchenQual","SaleCondition","LandSlope"]
important_features = num_features + cat_features

df2 = df2[important_features]
```

```python
for i in num_features:
    print(i)
```

**3. Models**

```python
# Spliting data
X = df2.drop("SalePrice", axis=1)
Y = df2["SalePrice"]
```

```python
# one-code encoding
X = pd.get_dummies(X, columns=cat_features)
```

```python
# Standardizing data
num_features.remove("SalePrice")
standard_scaler = StandardScaler()
X[num_features] = standard_scaler.fit_transform(X[num_features])
X.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=121)
```

***2.1. Lasso***

- Before applying hyperparameter tuning

```python
# Define the model
model_before = make_pipeline(Lasso(random_state=42))

# Fit the model
model_before.fit(X_train, Y_train)

# Make predictions on new data
predictions_ls = model_before.predict(X_test)
```

```python
#Print result
mae, mse, rmse, r_squared = evaluate(Y_test, predictions_ls)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r_squared)
print("-"*30)
rmse_cross_val = rmse_cv(model_before)
print("RMSE Cross-Validation:", rmse_cross_val)
```

-  Performs hyperparameter tuning using grid search on a Lasso regression model. After finding the best hyperparameters, we will evaluate the model's performance on a test set and calculates several metrics to assess the model's quality using the helper function above.

```python
# Define the model
model_after = make_pipeline(Lasso(random_state=42))

# Define the parameter grid
param_grid = {'lasso__alpha': [0.1, 0.5, 1.0]}

# Create the grid search object
grid_search = GridSearchCV(model_after, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the grid search to the entire dataset
grid_search.fit(X_train, Y_train)

# Access the best hyperparameters
best_params = grid_search.best_params_
best_alpha = best_params['lasso__alpha']

# Make predictions on new data using the best model
best_model = grid_search.best_estimator_
predictions_ls_after = best_model.predict(X_test)
```

```python
#Print result
mae, mse, rmse, r_squared = evaluate(Y_test, predictions_ls_after)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r_squared)
print("-"*30)
rmse_cross_val = rmse_cv(model_after)
print("RMSE Cross-Validation:", rmse_cross_val)
```

- Mean Absolute Error (MAE): The MAE value of 20077.75 indicates that, on average, the predicted values are approximately $20,077.75 away from the actual values.

- Mean Squared Error (MSE): The MSE value of 1011530001.10 indicates that, on average, the squared difference between the predicted and actual values is approximately 1,011,530,001.10. 

- Root Mean Squared Error (RMSE): The RMSE value of 31804.56 indicates that, on average, the predicted values have an error of approximately $31,804.56 compared to the actual values. 

- R2 Score: The R2 score value of 0.835 indicates that approximately 83.5% of the variance in the target variable is explained by the model. A higher R2 score suggests a better fit, where 1.0 represents a perfect fit.

- RMSE Cross-Validation: The RMSE value of 35922.77 indicates the average RMSE across the different folds in the cross-validation process. It gives an indication of the model's expected performance on unseen data.

- Overall, the evaluation results suggest that the Lasso regression model with the tuned hyperparameters performs reasonably well.
